#### - Issue 2

Most of Loan Date column is empty; we can turn the empty column to NaN, and convert the ones that aren't empty to international date standard. I'm thinking of dropping the column completely, but eorking with date of loaned players might be useful when performing analysis.

```python
#taking a look
clean_data['Loan Date End'].value_counts()
```

```python
#creating a function...
def change_to_int_date(df, col):
    '''This function changes dates to non-standard version to standard'''
    #creating arrays of month...
    month = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
    month2 = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

    #extracting the date...
    month_array = np.array(df[col].str.extract('(^.{0,3})'))

    day_array =  np.array(df[col].str.extract('(\d{1,2})'))
    day_array = np.where(pd.isna(day_array), 'None', day_array)
    day_array = day_array.flatten()

    year_array = np.array(df[col].str.extract('(\d{4})'))
    year_array = np.where(pd.isna(year_array), 'None', year_array)
    year_array = year_array.flatten()

    #converting month_array to number...
    new = []

    for i in month_array:
        if i in month2:
            new.append(month[month2.index(i)])
        else:
            new.append('None')

    new_month = np.array(new)

    #creating new date 
    new_date = year_array + '-' + new_month + '-' + day_array

    #replacing the column with the new values
    df[col] = new_date
    df[col] = df[col].replace('None-None-None', np.NaN, regex=True)
    df[col] = pd.to_datetime(df[col])
    return df[col]
```

```python
clean_data['Loan Date End'] = change_to_int_date(clean_data, 'Loan Date End')
```

```python
#testing our code
clean_data['Loan Date End'].value_counts()
```

Our date is now in the standard format.

#### - Issue 3
For 'Value', 'Wages', and 'release clause' column, they need to be only in figures; Euro and M/K should be removed.

```python
#creating a function to reduce repititive codes...
def change_to_num(df, col):
    '''This function changes the shortened values in a column to their proper numerical value'''
    #removing the Euro sign...
    df[col] = df[col].str.replace('€', '').astype(str)
    empty_list = []

    #converting the values to numbers
    for i in df[col]:
        if 'M' in i:
            num = float(i[:-1]) * 1000000
        elif 'K' in i:
            num = float(i[:-1]) * 1000
        else:
            num = float(i)
        empty_list.append(num)
    empty_list = np.array(empty_list)

    #updating our dataframe...
    df[col] = empty_list
    return df[col]
```

```python
#cleaning the 'Value' column...
clean_data['Value'] = change_to_num(clean_data, 'Value')
```

```python
#cleaning the 'Wages column...
clean_data['Wage'] = change_to_num(clean_data, 'Wage')
```

```python
#cleaning 'Release Clause column..
clean_data['Release Clause'] = change_to_num(clean_data, 'Release Clause')
```

#### - Issue 5
Star signal after values under 'W/F', 'SM', and 'IR'. Athough I am still contemplating leaving it as a categorical variable or numerical, probably we resort to changing it to a numerical variable.

```python
#creating a function...
def remove_star(df, col):
    '''This function replaces the star character with an empty string and turns it to int'''
    df[col] = df[col].str.replace('★', '').astype(int)
    return df[col]
```

```python
#removing star for 'W/F'
clean_data['W/F'] = remove_star(clean_data, 'W/F')

#removing star for 'SM'
clean_data['SM'] = remove_star(clean_data, 'SM')

#removing star in 'IR'
clean_data['IR'] = remove_star(clean_data, 'IR')
```

#### -Issue 6
Repeating '\n' in the 'club and contract' column should be removed, it is also necessary to separate the 2 column, having more than 1 info in a column violates tidiness.

```python
#replacing the '\n' with empty string...
clean_data['Team & Contract'] = clean_data['Team & Contract'].str.replace('\n', '').astype(str)
```

```python
#checking if it is gone
clean_data[['Team & Contract']].sample(5)
```

Now, we can separate the column.

```python
#splitting the column...
clean_data['Team'] = clean_data['Team & Contract'].str.extract(r'(\D+)')
month3 = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
new3 = []

#running a for loop to extract the team names
for i in clean_data['Team']:
    if i[-3:] in month3:
        new3.append(i[0:-3])
    else:
        new3.append(i)

clean_data['Team'] = new3

new4 = []

#running a for loop to extract contacts
for i in clean_data['Team & Contract']:
    if 'Loan' in i:
        new4.append('Loan')
    else:
        new4.append(i[-11:])

clean_data['Contract'] = new4
clean_data['Team'] = clean_data['Team'].str.replace(r'(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)', '')

#dropping 'Team & Contract' column...
clean_data.drop('Team & Contract', axis=1, inplace=True)
```

To explain what I did above, the team column I extracted above are coming from two kinds of contract; loan contracts and normal contracts. The teams with loan contract still contiains the month their contract is ending, so I had to run a loop to drop the month from the names.As for the loan column, I ran a conditional loop, so it can assign 'Loan' for those that are on loan, and their contract span for those that are not. I ignored the date for loaned contracts, as it is exactly same as in the loan end date column.

```python
#creating a function that converts feet to cm
def feet_to_cm(x):
    feet, inches = x.split("'")
    total_inches = int(feet)*12 + int(inches[:-1])
    cm = total_inches * 2.54
    return cm
#applying the function to the height column...
clean_data['Height'] = clean_data['Height'].apply(feet_to_cm)
```
